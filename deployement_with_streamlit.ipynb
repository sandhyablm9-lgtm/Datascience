```python
%%writefile diabeticapp.py

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st

# Loading the Dataset
df=pd.read_csv("diabetes.csv")

#Top Five Rows
df.head()

#Finding null values
df.isnull().sum()

# stats _analysis
df.describe()


df.info()

df.shape


#setting the target variable:
#outcome
df["Outcome"].value_counts()


#grouping the data
df.groupby("Outcome").mean()
# 0>>> non-diabetic patient
# 1>>> Diabetic patient


#Data Visualization
df.groupby("Outcome").mean().plot(kind="bar")

df.groupby("Outcome")["Pregnancies"].mean().plot(kind="bar")

df.pivot_table(index="Outcome",columns="Pregnancies",values="Age",aggfunc="mean").fillna(0)

x=df.drop("Outcome",axis=1)
x.head()

y=df["Outcome"]
y.head()

# Training and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=9,stratify=y)


#Shape of training and testing data
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)

# knn, svm
from sklearn import svm
classifier=svm.SVC(kernel="linear")  #linear(0,1)
classifier.fit(x_train,y_train)


#model prediction
prediction=classifier.predict(x_test)
prediction

from sklearn.metrics import accuracy_score
acc_score=accuracy_score(y_test,prediction)
acc_score

array=np.array([5,175,72,19,166,25.0,57,62])
reshaped_array=array.reshape(1,-1)
reshaped_array

prediction_array=classifier.predict(reshaped_array)
if prediction_array[0]==0:
    print("Non-Diabetic")
else:
    print("Diabetic")


st.title("Diabetes Prediction for no.of pregnancies")
#getting input from user:

pregnancies=st.text_input("Number of Pregnancies")
glucose=st.text_input("Enter the Glucose Level")
blood_pressure=st.text_input("Enter the Blood Pressure Value")
skin_thickness=st.text_input("Enter the Skin Thickness Value")
insulin=st.text_input("Enter the Insulin levels")
BMI=st.text_input("Enter the BMI Value")
Diabetespd=st.text_input("Enter the Diabetes Pedigree Function Value")
age=st.text_input("Enter the Age of the Person")

diagnosis=" "
if st.button("Diabetes Test Result"):
  diagnosis=classifier.predict([[pregnancies,glucose,blood_pressure,skin_thickness,insulin,BMI,Diabetespd,age]])

st.success(diagnosis)
```
